```python
#Problem 2
# Sort the avocado dataset by AveragePrice. Print the 8 rows with the largest values of AveragePrice
#wrong
#avo_max.head(8)

prices_avo = avo_data.sort_values('AveragePrice')
avo_max_8 = prices_avo.tail(8)
print(avo_max_8)
```

```python
#Problem 3
# Group the data by region and print the median AveragePrice for each region
median_price = avo_data.groupby('region').AveragePrice.median()
print(median_price)
```

```python
#Problem 4
# Group the data by region and print the sum of the "Total Volume" column for each region during the year 2017

year_2017 = avo_data[avo_data['year'] == 2017]
mean_volume = year_2017.groupby('region')['Total Volume'].mean()
count_volume = year_2017.groupby('region')['Total Volume'].count()
total_volume = mean_volume * count_volume
print(total_volume)
```

```python
#Problem 5
# Create a new column in the dataset called "Total Earnings" which is the "AveragePrice" times the
# "Total Volume" in each row

avo_data['Total Earnings'] = avo_data['AveragePrice'] * avo_data['Total Volume']
print(avo_data[['AveragePrice', 'Total Volume', 'Total Earnings']])
```

```python
#Problem 6
# Plot the average price of avocados vs time for "organic" avocados in Atlanta
# NOTE: Double check your dates on the x-axis. You may need to sort the data by "Date"
# first before plotting to get it to display correctly for whatever reason


plot_data = avo_data[(avo_data.region == 'Atlanta') & (avo_data.type == 'organic')]
plot_data = plot_data.sort_values('Date')
plot_data.plot(
    x ='Date',
    y ='AveragePrice',
    title='Average Price of Organic Avocados vs Time in Atlanta',
    xlabel='Date',
    ylabel='Average Price',
    kind='line'
)
plt.gcf().autofmt_xdate()
plt.show()
```

```python
#Problem 7
# Make a plot of AveragePrice vs year with error bars. To do this, find the mean AveragePrice in each year
# and use that for the y-values. Find the standard deviation of AveragePrice in each year and use that as 
# the y-error. Make sure that, kind="scatter"
# NOTE: There should only be 4 data points on your graph (one for each year).


mean_prices = avo_data.groupby('year').AveragePrice.mean()
std_prices = avo_data.groupby('year').AveragePrice.std()
years = mean_prices.index
mean_values = mean_prices.values
std_values = std_prices.values

plt.errorbar(years, mean_values, yerr=std_values, fmt='o')
plt.title('Average Price vs Year')
plt.xlabel('Year')
plt.ylabel('Average Price') 
plt.show()
```
